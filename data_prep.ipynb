```python
player_stats_2021_url = 'https://www.nbastuffer.com/2021-2022-nba-player-stats/'

player_stats_urls = ["https://www.nbastuffer.com/2020-2021-nba-player-stats/", 
                        'https://www.nbastuffer.com/2019-2020-nba-player-stats/',
                        'https://www.nbastuffer.com/2018-2019-nba-player-stats/',
                        'https://www.nbastuffer.com/2017-2018-nba-player-stats/']

player_stats_urls2 = ["https://www.nbastuffer.com/2016-2017-nba-regular-season-player-stats/", 
                        'https://www.nbastuffer.com/2015-2016-nba-regular-season-player-stats/',
                        'https://www.nbastuffer.com/2014-2015-nba-regular-season-player-stats/',
                        'https://www.nbastuffer.com/2013-2014-nba-regular-season-player-stats/']                        

salary_cap_url = 'https://www.spotrac.com/nba/cba/'

player_salaries_urls = ['https://hoopshype.com/salaries/players/',
                    "https://hoopshype.com/salaries/players/2020-2021/", 
                    'https://hoopshype.com/salaries/players/2019-2020/',
                    'https://hoopshype.com/salaries/players/2018-2019/',
                    'https://hoopshype.com/salaries/players/2017-2018/',
                    'https://hoopshype.com/salaries/players/2016-2017/',
                    'https://hoopshype.com/salaries/players/2015-2016/',
                    'https://hoopshype.com/salaries/players/2014-2015/',
                    'https://hoopshype.com/salaries/players/2013-2014/']
```

# Player Stats

```python
#Latest pages have two tables, need to grab only regular season
another_df_list = []

print(player_stats_2021_url)
r = requests.get(player_stats_2021_url)
df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
salaries_df = df_list[0] #no playoffs on this page, just current regular season
salaries_df.insert(1, 'Year', 2021 )
another_df_list.append(salaries_df)

for i, url in enumerate(player_stats_urls):
    print(url)
    r = requests.get(url)
    df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
    salaries_df = df_list[1] #this 1 grabs the second table on page - regular season, 0 would get playoffs
    salaries_df.insert(1, 'Year', 2020 - i)
    another_df_list.append(salaries_df)
```

```python
stats_df = pd.concat(another_df_list)
stats_df
```

```python
#Rename the mess that is these column names.  

stats_df_rename_dict = {'MIN%Minutes PercentagePercentage of team minutes used by a player while he was on the floor': 'MIN%',
       'USG%Usage RateUsage rate, a.k.a., usage percentage is an estimate of the percentage of team plays used by a player while he was on the floor': 'USG%',
       'TO%Turnover RateA metric that estimates the number of turnovers a player commits per 100 possessions': 'TO%',
       'eFG%Effective Shooting PercentageWith eFG%, three-point shots made are worth 50% more than two-point shots made. eFG% Formula=(FGM+ (0.5 x 3PM))/FGA' : 'eFG%',
       'TS%True Shooting PercentageTrue shooting percentage is a measure of shooting efficiency that takes into account field goals, 3-point field goals, and free throws.': 'TS%',
       'PPGPointsPoints per game.': 'PPG',
       'RPGReboundsRebounds per game.' : 'RPG',
       'TRB%Total Rebound PercentageTotal rebound percentage is estimated percentage of available rebounds grabbed by the player while the player is on the court.' : 'TRB%', 
       'APGAssistsAssists per game.' : 'APG',
       'AST%Assist PercentageAssist percentage is an estimated percentage of teammate field goals a player assisted while the player is on the court': 'AST%',
       'SPGStealsSteals per game.': 'SPG',
       'BPGBlocksBlocks per game.': 'BPG',
       'TOPGTurnoversTurnovers per game.': 'TOPG',
       'VIVersatility IndexVersatility index is a metric that measures a playerâ€™s ability to produce in points, assists, and rebounds. The average player will score around a five on the index, while top players score above 10': 'VI',
       'ORTGOffensive RatingIndividual offensive rating is the number of points produced by a player per 100 total individual possessions.': 'ORTG',
       'DRTGDefensive RatingIndividual defensive rating estimates how many points the player allowed per 100 possessions he individually faced while staying on the court.': 'DRTG',
       'Tor%Turnover RateA metric that estimates the number of turnovers a player commits per 100 possessions': 'Tor%',
       'TOrTurnover RateA metric that estimates the number of turnovers a player commits per 100 possessions': 'TOr'}

stats_df.rename(columns=stats_df_rename_dict, inplace=True)
```

# Older Data
Doesn't have all advanced metrics.  Maybe better to not include? 

```python
another_df_list2 = []

for i, url in enumerate(player_stats_urls2):
    print(url)
    r = requests.get(url)
    df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
    salaries_df = df_list[0]
    salaries_df.columns = salaries_df.columns.droplevel(level = 0)
    salaries_df.insert(1, 'Year', 2016 - i)
    another_df_list2.append(salaries_df)
```

```python
stats_df2 = pd.concat(another_df_list2)
stats_df2
```

```python
#Need to relabel a few of the columns
old_stats_rename_dict = {'PLAYER':'FULL NAME',
                        'Rank':'RANK'}

stats_df2.rename(columns=old_stats_rename_dict, inplace=True)

#Convert Turn over percentage to match new data which is turnovers over 100 possessions
#stats_df2['TO%'] = stats_df2['TO%'] * 100
```

```python
total_stats_df = pd.concat([stats_df, stats_df2])
```

```python
#Convert Turn over percentage to match new data which is turnovers over 100 possessions
total_stats_df['TOr'] = total_stats_df['TOr'] * 100

#Combine the 3 turnover per 100 game columns into one

total_stats_df['TO_100_Games'] = total_stats_df['TO%'].combine_first(total_stats_df['Tor%'].combine_first(total_stats_df['TOr']))
```

```python
#Drop the extra turnover columns
total_stats_df.drop(['TOr', 'Tor%', 'TO%', 'TOPG'], axis = 1, inplace=True)
```

```python
#Check out NA Values
total_stats_df.isna().sum(axis=0).sort_values(ascending = False)[0:10]
```

```python
#Drop the columns shown
total_stats_df.drop(['RANK', 'Unnamed: 25_level_1'], axis = 1, inplace=True)
```

```python
#Fill in advanced stats

#Effective Field Goal Percentage Formula=[(All Field Goals Made) + 0.5*(3P Field Goals Made)]/(All Field Goal Attempts)
efg_idx = total_stats_df['eFG%'].isna()

total_stats_df['eFG%'] = (total_stats_df['2PA'] * total_stats_df['2P%'] + 
                        0.5 * (total_stats_df['3PA'] * total_stats_df['3P%'])) / \
                        (total_stats_df['2PA'] + total_stats_df['3PA'])

#nan values are people with no shots.  0% eFG%
total_stats_df['eFG%'].fillna(0, inplace = True)                      

#nan values are people with no turnovers
total_stats_df['TO_100_Games'].fillna(0, inplace = True)   

#nan values are people with no shots
total_stats_df['TS%'].fillna(0, inplace = True)   

total_stats_df['AGE'].fillna(28, inplace = True)   
```

```python
total_stats_df.isna().sum(axis=0).sort_values(ascending = False)[0:3]
```

```python
total_stats_df.to_csv('data/stats_only-2013-2021.csv', index = False)
```

# Salaries

```python
another_df_list = []

year_strings = ['2021/22', '2020/21', '2019/20', '2018/19', '2017/18', '2016/17', '2015/16', '2014/15', '2013/14'] #used to consolidate salaries into one column

for i, url in enumerate(player_salaries_urls):
    print(url)
    r = requests.get(url)
    df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
    salaries_df = df_list[0]
    salaries_df.insert(1, 'Year', 2021 - i)
    salaries_df['Salary'] = salaries_df[year_strings[i]]
    another_df_list.append(salaries_df)
    
```

```python
salaries_df = pd.concat(another_df_list)

#drop the extra columns
salaries_df = salaries_df[['Player', 'Year', 'Salary']]

#Clean up the $ signs and commas, change type to int
salaries_df['Salary'] = salaries_df['Salary'].str.replace("[^0-9.]", '').astype(int)
salaries_df.reset_index(drop=True, inplace=True)
salaries_df
```

```python
salaries_df.to_csv('data/salaries_only.csv', index = False)
```

# Salary Cap

```python
r = requests.get(salary_cap_url)
df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
salary_cap_df = df_list[0]

#remove $ and commas
salary_cap_df['Cap Maximum'] = salary_cap_df['Cap Maximum'].str.replace("[^0-9.]", '').astype(int)

salary_cap_df.head()
```

```python
salary_cap_df.to_csv('data/salary_cap.csv', index = False)
```

# Merge Large Dataset

```python
#Merge Stats and Salary
merged_df = total_stats_df.merge(salaries_df, how = 'left', left_on = ['FULL NAME', 'Year'], right_on=['Player', 'Year'])
merged_df.drop('Player', axis = 1, inplace=True)
```

```python
#Merge Stats, Salary, and Salary Cap
merged_df = merged_df.merge(salary_cap_df[['Year', 'Cap Maximum']], how = 'left', on = 'Year')

#Add Players % of Cap
merged_df['Salary%OfCap'] = round(merged_df['Salary'] / merged_df['Cap Maximum'], 4)
```
